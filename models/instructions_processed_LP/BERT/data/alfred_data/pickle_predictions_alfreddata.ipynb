```python
with open('val_unseen_text_with_ppdl_low_appended.p', 'rb') as file:
    testing = pickle.load(file)

print(len(testing['x_low']))
count = {}
repeated = 0
repeated_list = []

# count the number of repeated keys in the appended data
for x in testing['x']:
    if x in count.keys():
        count['x'] += 1
        repeated += 1
        repeated_list.append(x)
    else:
        count['x'] = 1

testing
```
